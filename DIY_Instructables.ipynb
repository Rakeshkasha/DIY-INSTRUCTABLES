```
web_page ="https://www.instructables.com/"
html = url.urlopen(web_page).read()
title=[]
page=[]
listoflinks=[]
#web={"names":[],"links":[]}
#web = {'name':[],'link':[]}
web = {'title':[],'link':[]}
soup = BeautifulSoup(html, 'html.parser')
for titles in soup.find_all("span",{"class":"anchor-text"}):
  title.append(titles.text)
for links in soup.find_all("a",{"class":"home-content-explore-link"}):
  page.append(links["href"])
print(title)
print(page)
web_link=[]
for i in page:
  web_link.append("https://www.instructables.com/"+i)
print(web_link)
for i in title:
  web["title"].append(i)
for j in web_link:
  web["link"].append(j)
```
